# Two options for subplots


```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                            ncols=2, 
                                            figsize=(10, 5))

#plot to each different axis

ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2, 
                       ncols=2, 
                       figsize=(10, 5))

# Plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

### Which one should you use? (Pyplot or Matplotilb OO method)?

* When plotting something quickly, it is ok to use Pyplot method
* When plotting something more advance, use Matplotlib OO method

```python
over_50 = heart_disease[heart_disease["age"] >= 50]
over_50.head()
# len(over_50)
```

```python
# Pyplot method

over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target");
```

```python
# OO Method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
ax.set_xlim([45, 100])
ax.set_title("Age and Cholesterol Levels")
```

```python
# OO Method from scratch
fig, ax  = plt.subplots(figsize=(10, 6))

#Plot the data
scatter = ax.scatter(x=over_50['age'], y=over_50['chol'], c=over_50["target"]);

#Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol",);

#Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle="--")
```

```python
over_50
```

```python
# Subplot of chol, age, thalach
plt.style.use('seaborn')

fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True)
#Plot the data for ax0
scatter0 = ax0.scatter(x = over_50['age'], 
                       y= over_50['chol'], 
                       c=over_50['target'])

# customize legends, labels and add a mean line
ax0.set(title="Heart Disease and Cholesterol Levels",
      ylabel="Cholesterol");

ax0.legend(*scatter.legend_elements(), 
           title="Target");

ax0.axhline(over_50['chol'].mean(), 
            linestyle="--")

#plot data for ax1
scatter1 = ax1.scatter(x = over_50['age'], 
                       y=over_50['thalach'], 
                       c=over_50['target'])

# customize with legends, labels and add a mean line
ax1.set(title="Heart Disease and Max Heart Rate", 
        xlabel="Age", ylabel="Thalach");

ax1.legend(*scatter1.legend_elements(), 
           title= "Target");
ax1.axhline(over_50['thalach'].mean(), 
            linestyle="--");

# Add title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
# OO Method from scratch
plt.style.use('seaborn-whitegrid')
fig, ax  = plt.subplots(figsize=(10, 6))

#Plot the data
scatter = ax.scatter(x=over_50['age'], 
                     y=over_50['chol'], 
                     c=over_50["target"],
                    cmap="winter"); # changes the color scheme

#Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol",);

#Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), 
           linestyle="--")
```

```python
# Customizng the x and y limitations

# Subplot of chol, age, thalach
plt.style.use('seaborn-whitegrid')

fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True)
#Plot the data for ax0
scatter0 = ax0.scatter(x = over_50['age'], 
                       y= over_50['chol'], 
                       c=over_50['target'],
                      cmap="winter")

# customize legends, labels and add a mean line
ax0.set(title="Heart Disease and Cholesterol Levels",
      ylabel="Cholesterol");

ax0.set_xlim([45, 80])
ax0.set_ylim([45, 600])

ax0.legend(*scatter.legend_elements(), 
           title="Target");

ax0.axhline(over_50['chol'].mean(), 
            linestyle="--");

#plot data for ax1
scatter1 = ax1.scatter(x = over_50['age'], 
                       y=over_50['thalach'], 
                       c=over_50['target'],
                      cmap="winter")

# customize with legends, labels and add a mean line
ax1.set(title="Heart Disease and Max Heart Rate", 
        xlabel="Age", ylabel="Thalach");
ax1.set_ylim([45, 200])

ax1.legend(*scatter1.legend_elements(), 
           title= "Target");
ax1.axhline(over_50['thalach'].mean(), 
            linestyle="--");

# Add title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
